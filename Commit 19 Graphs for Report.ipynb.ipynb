```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
### Create graphs for Report
```

```python
# CREATE R2 PLOT WITH RESULTS

r2_scores_ANN = [0.959439, 0.967420, 0.992664] 
r2_scores_RF = [.95, .93, .95]  
r2_scores_SVM = [-0.04, -0.04, -0.04]  


# Group names
group_names = ['10%', '20%', '30%']

# Bar positions
bar_positions = np.arange(len(group_names))

# Bar width
bar_width = 0.20

# Create grouped bar graph
plt.bar(bar_positions - bar_width, r2_scores_ANN, width=bar_width, label='ANN')
plt.bar(bar_positions, r2_scores_RF, width=bar_width, label='RF')
plt.bar(bar_positions + bar_width, r2_scores_SVM, width=bar_width, label='SVM')

# Customize the plot
plt.xlabel('Models and Percetnage Splits', weight='bold', fontsize=12)
plt.ylabel('R2 Score', weight='bold', fontsize=12)
plt.title('R2 Scores for Percentage Splits', weight='bold', fontsize=16)
plt.xticks(bar_positions, group_names)
plt.legend()

# Place legend outside the plot
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```
