```python
# This shows the correaltion between labour force and population growth, each dot represents. sepcified region.
# For Africa eastern and southern as population grows so does labour force quite well but no so much in other regions.
# This can help give us an insight into the contributing factors for these indicators.

import pandas as pd
import matplotlib.pyplot as plt

# Read in the data
df = pd.read_csv('WDI_Africa_5k.csv')

# Define the country list
country_list = ['Central African Republic', 'South Africa', 'Africa Western and Central','Africa Eastern and Southern']

# Filter the data by country and indicator
df_filtered = df[(df['Country Name'].isin(country_list)) & (df['Indicator Name'].isin(['Labor force, total', 'Population, total']))]

# Pivot the data to get separate columns for each indicator
df_pivot = df_filtered.pivot(index='Country Name', columns='Indicator Name', values='2019')

# Create the scatter plot
plt.scatter(df_pivot['Population, total'], df_pivot['Labor force, total'])

# Add axis labels and title
plt.xlabel('Population, total')
plt.ylabel('Labor force, total')
plt.title('Comparison of Labor force and Population in African Regions')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data
df = pd.read_csv('WDI_Africa_5k.csv')

# Define the country list
country_list = ['Central African Republic', 'South Africa', 'Africa Western and Central','Africa Eastern and Southern']

# Filter the data by country and indicator
df_filtered = df[(df['Country Name'].isin(country_list)) & (df['Indicator Name'].isin(['Labor force, total', 'Population, total']))]

# Pivot the data to get separate columns for each indicator
df_pivot = df_filtered.pivot(index='Country Name', columns='Indicator Name', values='2019')

# Create the scatter plot
plt.scatter(df_pivot['Population, total'], df_pivot['Labor force, total'])

# Add axis labels and title
plt.xlabel('Population, total')
plt.ylabel('Labor force, total')
plt.title('Comparison of Labor force and Population in African Regions')

# Add a legend
for i, txt in enumerate(df_pivot.index):
    plt.annotate(txt, (df_pivot['Population, total'][i], df_pivot['Labor force, total'][i]))
    
# Show the plot
plt.show()
```

```python
# In this code we are using the imputaton method One Hot Encoding.
# We use this as we feel it is best suited to our data frame as it handles both categorical and numerical values well.
# TThis method is great at handling data set variable that have complex and no direct linear correlation as whole.

import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

# Read in the data
df = pd.read_csv('WDI_Africa_5k.csv')

# Define the categorical and numerical columns
cat_cols = ['Country Name', 'Indicator Name', 'Indicator Code', 'Country Code']
num_cols = list(set(df.columns) - set(cat_cols))

# Impute the missing values in numerical columns using mean imputation
imputer = SimpleImputer(strategy='mean')
df[num_cols] = imputer.fit_transform(df[num_cols])

# One hot encode the categorical columns
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
cat_data = encoder.fit_transform(df[cat_cols])
cat_cols_encoded = encoder.get_feature_names(cat_cols)

# Combine the numerical and encoded categorical data
df_encoded = pd.DataFrame(cat_data, columns=cat_cols_encoded)
df_encoded[num_cols] = df[num_cols]

# Show the encoded dataframe
print(df_encoded.head())
```
